```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

# make sure the plot is displayed in this notebook
%matplotlib inline
# specify the size of the plot
plt.rcParams['figure.figsize'] = (16, 6)

# for auto-reloading extenrnal modules
%load_ext autoreload
%autoreload 2
```

```python
N = 20
w1 = .62831853
a = 0
w2 = 0
b = 0
n = np.linspace(0, N-1, N)

dn = np.random.default_rng().normal(0, 1, size=N)
xn = (np.cos(w1*n) + a * np.cos(w2*n) + b*dn)
xn_fft = np.fft.fft(xn.astype(complex))

plt.ylabel("Magnitude of DFT of Case 1")
plt.xlabel("k")
plt.stem(np.abs(xn_fft))
```
