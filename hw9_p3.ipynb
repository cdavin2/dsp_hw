Part C

```python
U_50000 = np.random.uniform(0, 1, size = 50000)
X_50000 = 3 * U_50000**(1/3) ## fXx
quant = []

for i in range(len(X_50000)):
  if X_50000[i] < 1.5:
    quant.append(.75)
  else:
    quant.append(2.25)

error = np.sum( (quant-X_50000)**2 ) / 50000
print(error)
```

Part D

```python
U_50000 = np.random.uniform(0, 1, size = 50000)
X_50000 = 3 * U_50000**(1/3) ## fXx
quant = []

for i in range(len(X_50000)):
  if X_50000[i] < 2.07:
    quant.append(1.55)
  else:
    quant.append(2.59)

error = np.sum( (quant-X_50000)**2 ) / 50000
print(error)
```

The rms error of the quantizer with optimal values is less than the rms error of the quantizer with non-optimal values. This  makes sense as we would expect a lower error when using optimal values compared when not.
