```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

# make sure the plot is displayed in this notebook
%matplotlib inline
# specify the size of the plot
plt.rcParams['figure.figsize'] = (16, 6)

# for auto-reloading extenrnal modules
%load_ext autoreload
%autoreload 2
```

```python
n = np.linspace(-5, 20, 26)
xna_0 = (0 <= n).astype(float) - (6 <= n).astype(float)
xna_1 = (1 <= n).astype(float) - (7 <= n).astype(float)
xna_2 = (2 <= n).astype(float) - (8 <= n).astype(float)

yna1 = np.zeros(len(xna_0)) # len(xna) is 26
for i in range(len(xna_0)): 
  yna1[i] = (1/4) * (xna_0[i] - 2*xna_1[i] + xna_2[i])

yna2 = np.zeros(len(xna_0)) # len(xna) is 26
yna2[0] = 0
for i in range(1, len(xna_0)): # 1...20
  yna2[i] = xna_0[i] + xna_1[i] + yna2[i-1]

#######################################################

n = np.linspace(-5, 20, 26)
xnb_0 = np.cos((np.pi*n)/5) * (0 <= n).astype(float) 
xnb_1 = np.cos((np.pi*(n-1))/5) * (1 <= n).astype(float) 
xnb_2 = np.cos((np.pi*(n-2))/5) * (2 <= n).astype(float) 

ynb1 = np.zeros(len(xnb_0)) # len(xna) is 26
for i in range(len(xna_0)): 
  ynb1[i] = (1/4) * (xnb_0[i] - 2*xnb_1[i] + xnb_2[i])

ynb2 = np.zeros(len(xnb_0)) # len(xna) is 26
ynb2[0] = 0 # initial condition 
for i in range(1, len(xnb_0)): # 1...20
  ynb2[i] = xnb_0[i] + xnb_1[i] + ynb2[i-1]

#######################################################

fig, axs = plt.subplots(3, 2)
plt.gcf().set_size_inches(20, 20)
axs[0, 0].stem(xna_0)
axs[0, 0].set_title("x[n] for Part A of HW1: Problem 3")
axs[1, 0].stem(yna1)
axs[1, 0].set_title("y1[n] for Part A of HW1: Problem 3")
axs[2, 0].stem(yna2)
axs[2, 0].set_title("y2[n] for Part A of HW1: Problem 3")

axs[0, 1].stem(xnb_0)
axs[0, 1].set_title("x[n] for Part B of HW1: Problem 3")
axs[1, 1].stem(ynb1)
axs[1, 1].set_title("y1[n] for Part B of HW1: Problem 3")
axs[2, 1].stem(ynb2)
axs[2, 1].set_title("y2[n] for Part B of HW1: Problem 3")
fig.show()
```
